```python
import seaborn as sns
```

```python
# Loading datasets from seaborn
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# (4) Categorical Plots
# (i) Categorical Scatter Plot --> We can draw Scatter plot using the categorical column also but the second column must be a numerical one

# (a) Strip Plot :

tips # using tips dataset , one(total_bill) i.e. numerical and the second(day) i.e. categorical column
```

```python
sns.stripplot(data=tips, x='day', y='total_bill') # here stipplot() is an axes level function
# here the seaborn automatically assigns some noise to the graph to enhance the visibility
```

```python
# to remove the noise we can do this (Using jitter parameter)
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
# Using figure level function (catplot())
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
# Increasing the jitter value

sns.catplot(data=tips, x='day', y='total_bill', kind='strip', jitter=1.5)
```

```python
# Using hue parameter :
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', hue='sex')
```

```python
# (b) Swarm Plot
# Describes the distribution of Data and shows how the data is divided

sns.catplot(data=tips, x='day', y='total_bill', kind='swarm') # using figure level function(catplot)
```

```python
# using the axes level function
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
# using hue parameter
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# (ii) Categorical Distribution Plots

# (a) Box Plot : very important plot and explains various details
# The Outliers are the values which exceed the calculated minimum and maximum values
# The minimum and maximum are not the min/max values of the whole data, they're calculated.
```

```python
# syntax --> sns.boxplot(data, x(categorical col), y(numerical col))
# Example :
sns.boxplot(data=tips, x='sex', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill') # box plot ke help se sabhi categories ka max/min/median/IQR calculate kiya jaa skta h side-by-side
```

```python
# using figure level function :
sns.catplot(data=tips, kind='box', x='sex', y='total_bill')
```

```python
# hue parameter :
sns.catplot(data=tips, kind='box', x='day', y='total_bill', hue='sex')
```

```python
# single boxplot -> numerical col

sns.catplot(data=tips, kind='box', y='total_bill')
```

```python
# (b) Violin Plot :

sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
# using figure level functions --> catplot()
sns.catplot(data=tips, kind='violin', x='day', y='total_bill')
```

```python
# hue parameter
sns.catplot(data=tips, kind='violin', x='day', y='total_bill', hue='sex')
```

```python
# split parameter : to avoid creation of side by side plots for the hue parameter

sns.catplot(data=tips, kind='violin', x='day', y='total_bill', hue='sex', split=True)
```

```python
# (iii) Categorical Estimate Plots : used to calculate the mean/median/avg of a categorical col VS. numerical col
# (a) Bar Plot
# using axes level function
sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
# using figure level function
sns.catplot(data=tips, kind='bar', x='sex', y='total_bill')
```

```python
# using hue parameter :
sns.catplot(data=tips, x='sex', y='total_bill', hue='smoker', kind='bar')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', errorbar=('ci',False)) #(OR) errorbar=None
```

```python
# estimator Parameter : (estimator = 'mean'/ 'median'/ np.std)
import numpy as np
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.std)
```

```python
# (b) Point plot --> it is completely same to the bar plot except the bars, lines are only drawn
# The lines show the difference between the plots

sns.catplot(data=tips, kind='point', x='sex', y='total_bill', hue='smoker', estimator='mean', errorbar = ('ci', False))
```

```python
# (c) Count Plot
# A special case for bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable.
# This is similar to histogram over a categorical, rather than a quantative/numerical, variable.
# Its working is same as the Pandas value count function
```

```python
# count plot
sns.countplot(data=tips, x='sex')
# using axes level function
```

```python
sns.catplot(data=tips, kind='count', x='day', hue='sex')
# using figure level function
```

```python
# Faceting using catplot
# (a) for countplot
sns.catplot(data=tips, kind='count',x='day', col='sex')
```

```python
# faceting using catplot
# (b) boxplot

sns.catplot(data=tips, kind='box',x='sex', y='total_bill',col='day', col_wrap=2)
```

```python
# (5) Regression Plots
# regplot --> axes level function
# lmplot --> figure level function
# In the simplest invocation, both functions draw a scatterplot of two variables,
# x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

# When we determine the mathematical relationship between 2 quantities , it is k/a "Regression"
# It draws a linear regression line in the plot, which gives the values of slope(m) and y intercept(c) which is used as a formula.
# the lighter colour shade along the straight line is the error.
```

```python
# axes level function (It doesn't has the hue parameter)

sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# figure level function
# no need of kind parameter
sns.lmplot(data=tips, x='total_bill', y='tip',hue='sex')
```

```python
# (6) Residual/Resid plot :

sns.residplot(data=tips, x='total_bill', y='tip')
```

```python
# (7) Multi Plots --> Mutli Grids
# (a) Facet Plot --> Facet Grids
# facets(rows, cols) as we used before in regplot, displot, catplot, lmplot

# Facet Grid is more customisable, flexible, and is better to know.
```

```python
# (a) Facet Plot --> Facet Grids
g = sns.FacetGrid(data=tips, row='time', col='day')
g.map(sns.stripplot, 'sex', 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, row='time', col='day', hue='smoker')
g.map(sns.stripplot, 'sex', 'total_bill')
g.add_legend()
```

```python
# Plotting Pairwise relationship (PairGrid Vs Pairplot)
# It only works on numerical cols
# It forms pairs of the data and then plots the graph
# It is used when we have 4-5-6-..-10 numerical cols
```

```python
# (i) Pairplot
```

```python
# Using IRIS dataset for this
iris #It has 4 numerical cols and the PairPLot makes pairs of 2 and
# plots it on a scatterplot for different numerical cols
# and plots a histogram for same numerical cols
```

```python
sns.pairplot(iris)
# The graphs on the opposite side of the diagonal is symmetrical
# The diagonal graphs are histogram (since have same numerical cols)
```

```python
# Hue Parameter

sns.pairplot(iris, hue='species')
```

```python
# (ii) PairGrid :
g = sns.PairGrid(data=iris)

#g.map()
g.map(sns.lineplot)
```

```python
# hue parameter

g = sns.PairGrid(data=iris, hue='species')
# g.map()
g.map(sns.scatterplot)
```

```python
# map_diag --> map_offdiag separately

g = sns.PairGrid(data=iris)
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag --> map_upper --> map_lower

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.add_legend()
```

```python
# vars (restricting the cols on which I want to work)

g = sns.PairGrid(data=iris, hue='species',vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
g.add_legend()
```

```python
# (c) JointGrid Vs Jointplot
# (i) Jointplot

sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
# kind parameter :
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist', hue='sex')
```

```python
# (ii) JointGrid :
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.kdeplot, sns.boxplot)
```
