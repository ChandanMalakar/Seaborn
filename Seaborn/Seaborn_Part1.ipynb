```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px # using this library for the toy datasets provided by it
```

```python
tips = sns.load_dataset('tips')
#seaborn has this predefined dataset in it
tips
```

```python
# scatter plot --> axis level function
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

```python
# the above same graph using fig level function (relplot shd be used)
# relplot --> figure level function --> square shape
# It is adviced to use fig level function more

sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='day', style='time')
```

```python
# line plot

gap = px.data.gapminder() # in the above plotly.express library there's one toy dataset known as gapminder
temp_df = gap[gap['country']=='India']
```

```python
# axis level function --> lineplot()

sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# figure level function --> relplot()

sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# Discrete bins are automatically set for categorical values

# ex : from the tips dataset, we want to know how many people came on the respective days

# It can't be called a histogram, instead we call it "CountPlot"

sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
# KDE Plot (a) Univariate
# rather than using discrete bins, a KDE Plot smooths the observation with a Guassian Kernel, producing a continuous density estimate
```

```python
tips #using tips dataset
```

```python
sns.kdeplot(data=tips, x='total_bill') # here in the below plot the Density is the probability/roughly calculated, not the exact value
```

```python
  # From the above plot, we can say that the probability of 20(total bill) is 5%=0.05
```

```python
# using it with the help of (displot) the figure level function

sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# alag alag dekhna ho toh (hue) use kr skte
# hue --> fill
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, height=10, aspect=2)
#The fill parameter is used similarly to the element->step to determine the area of the plot.

# The height and width of a figure level function can't be increased using the plt.figure(figsize=(10,10))
# that is for axes level func, therefore height and aspect parameters are used.
```

```python
# Rugplot

# Not commonly used, used as a support for other plots.
# Plots marginal distribution by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill') # In the below graph, the orange ticks are the rug plot which shows the density distribution area
```

```python
# (b) Bivariate histogram
# A bivariate histogram bins the data within the rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

# In the below graphs the histogram is drawn from both the axes x,y then, the histograms which intersect each other are filled with some color.

sns.displot(data=tips, x='total_bill', y='tip', kind='hist') # using the figure level function
```

```python
# (OR using the axes level function)
sns.histplot(data=tips, x='total_bill', y='tip')

# The darkest color means we have the most combinations for that intersection value
```

```python
# (b) Bivariate KDE Plot
# A bivariate kdeplot smoothes the (x,y) observations with a 2D gaussian

sns.kdeplot(data=tips, x='total_bill', y='tip')

# Lines closer to each other describes the higher density in that location (more number of customers are there) and vice versa for the opposite.
```

```python
# Matrix Plots
# (a) Heatmap
# (b) Clustermap

# (a) Heatmap : Plots rectangular data as a color-encoded matrix.
# It is an axes level function
```

```python
gap
```

```python
# har country ka har saal life expectancy kitna tha plot krna hai

# since the above data is in 'long' format, we need to convert it to 'wide' format using 'pivot()' function

temp_df = gap.pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python
# representing the above grid in the form of a heatmap
plt.figure(figsize=(10,10)) #used to resize the figure level function inside which the heatmap is there(done with the help of matplotlib library)
sns.heatmap(temp_df)
```

```python
# annotations in the heatmap
# plotting the same heatmap for only european countries

europe_df = gap[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')
europe_df
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(europe_df)
```

```python
# Similarly I will do it for asia

asia_df = gap[gap['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp')
asia_df
```

```python
sns.heatmap(asia_df)
```

```python
# Annotations (annot) --> sirf colors ni but uske saath lifeExp ki value bhi

plt.figure(figsize=(8,8))

# giving linewidth will increase the gaps between the boxes and it enhances the readability
sns.heatmap(asia_df, annot=True, linewidth=0.5)
```

```python
# colours of the heatmap can also be changed as per the cmap=(value) # here cmap stands for (Colour map)

plt.figure(figsize=(8,8))
sns.heatmap(asia_df, annot=True, linewidth=0.5, cmap='summer')
```

```python
# (b) Clustermap
# Plot a matrix dataset as a hierarchically-clustered heatmap.
# This function requires scipy to be available

# Importing the iris dataset from the plotly library(import plotly.express as px)
iris = px.data.iris() # This dataset is about flowers
iris
```
